```
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
img = cv2.imread('tracer.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
plt.imshow(img)
plt.grid(False)
plt.axis('off')
plt.show()
```

```
edge = cv2.Canny(img, threshold1=127, threshold2=200)
```

```
plt.imshow(edge)
plt.grid(False)
plt.axis('off')
plt.title('no manual blur and manuel thresholds')
plt.show()
```

```
median =  np.median(img)
```

```
upper = int(max(0, 0.7*median))
lower = int(min(0, 1.3*median))
```

```
edge = cv2.Canny(img, threshold1=lower, threshold2=upper+100)
```

```
plt.imshow(edge)
plt.grid(False)
plt.axis('off')
plt.title('no manual blur')
plt.show()
```

```
blurred_img = cv2.blur(img, ksize=(6,6))
```

```
edge = cv2.Canny(blurred_img, threshold1=lower, threshold2=upper+100)
```

```
plt.imshow(edge)
plt.title('with manual blur')
plt.grid(False)
plt.axis('off')
plt.show()
```

```
edge = cv2.Canny(blurred_img, threshold1=127, threshold2=160)
```

```
plt.imshow(edge)
plt.grid(False)
plt.axis('off')
plt.title('manual blur and manual thresholds')
plt.show()
```
